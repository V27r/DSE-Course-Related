### 3. For missing values find and replace with arithmetic mean of the attributes


```python
for x in [ 'calories', 'protein', 'fat', 'sodium', 'fiber','carbo', 'sugars', 'potass', 'vitamins']:
    df[x] = df[x].replace('?','-1')
    df[x] = df[x].astype('float64')
    df[x] = np.where((df[x]==-1), df[x].mean(), df[x])
z = []
for x in [ 'calories', 'protein', 'fat', 'sodium', 'fiber','carbo', 'sugars', 'potass', 'vitamins']:
    z.append((df[x]==(-1)).sum())
print(z)
df
```

### 4. Find and replace outliers with median.


```python
def remove_outlier_IQR(df):
    Q1=df.quantile(0.25)
    Q3=df.quantile(0.75)
    IQR=Q3-Q1
    df_final=df[~((df<(Q1-1.5*IQR)) | (df>(Q3+1.5*IQR)))]
    return df_final
df[[ 'calories', 'protein', 'fat', 'sodium', 'fiber','carbo', 'sugars', 'potass', 'vitamins']]=remove_outlier_IQR(df[[ 'calories', 'protein', 'fat', 'sodium', 'fiber','carbo', 'sugars', 'potass', 'vitamins']])
for x in [ 'calories', 'protein', 'fat', 'sodium', 'fiber','carbo', 'sugars', 'potass', 'vitamins']:
     df[x] = df[x].replace(np.nan,df[x].median())
```
