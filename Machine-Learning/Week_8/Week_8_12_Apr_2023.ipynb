### Write the python code to build the ensemble learning models on atleast 4 datasets (with and without using pipeline). Build the models with atleast 4 train_test_split combinations and different set of hyper-parameter values. Perform the data pre-processing and performance evaluation for the models.


```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,BaggingClassifier,AdaBoostClassifier,VotingClassifier
from sklearn.model_selection import train_test_split,cross_val_score, GridSearchCV
from sklearn.svm import SVC
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.pipeline import Pipeline
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
df=pd.read_csv('mushrooms.csv')
```

```python
for label in df.columns:
    df[label]=LabelEncoder().fit(df[label]).transform(df[label])
df.info()
```

```python
X=df.drop(['class'],axis=1)
Y=df['class']
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
model=DecisionTreeClassifier(criterion='entropy',max_depth=1)
model=model.fit(X_train,Y_train)
y_pred=model.predict(X_test)
print("Accuracy of model without ensembling:",metrics.accuracy_score(Y_test,y_pred)*100,"%")
```

```python
AdaBoost=AdaBoostClassifier(base_estimator=model,n_estimators=70,learning_rate=0.5)
boost_model=AdaBoost.fit(X_train,Y_train)
y_pred=boost_model.predict(X_test)
predictions=metrics.accuracy_score(Y_test,y_pred)
print("Accuracy of ensembled model:",predictions*100,"%")
```

```python
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier, BaggingClassifier, AdaBoostClassifier, VotingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import pandas as pd

# Define a function to run the classifiers with or without pipelining
def run_classifiers(pipelining):
    # Create an empty DataFrame to store the results
    results = pd.DataFrame()

    for n in range(1, 100, 10):
        # Define classifiers with different hyperparameters
        rf = RandomForestClassifier(n_estimators=n, max_depth=5, random_state=42)
        bag = BaggingClassifier(base_estimator=DecisionTreeClassifier(max_depth=5), n_estimators=n, random_state=42)
        ada = AdaBoostClassifier(base_estimator=DecisionTreeClassifier(max_depth=5), n_estimators=n, random_state=42)
        voting = VotingClassifier(estimators=[('rf', rf), ('bag', bag), ('ada', ada)], voting="hard")

        if pipelining:
            # Create a pipeline for each classifier
            rf = Pipeline(steps=[('classifier', rf)])
            bag = Pipeline(steps=[('classifier', bag)])
            ada = Pipeline(steps=[('classifier', ada)])
            voting = Pipeline(steps=[('classifier', voting)])

        # Fit classifiers to the training data
        rf.fit(X_train, Y_train)
        bag.fit(X_train, Y_train)
        ada.fit(X_train, Y_train)
        voting.fit(X_train, Y_train)

        # Predict on the test data and calculate accuracy
        rf_accuracy = accuracy_score(Y_test, rf.predict(X_test))
        bag_accuracy = accuracy_score(Y_test, bag.predict(X_test))
        ada_accuracy = accuracy_score(Y_test, ada.predict(X_test))
        voting_accuracy = accuracy_score(Y_test, voting.predict(X_test))

        # Append results for this value of n to the DataFrame
        results_n = pd.DataFrame({
            'Classifier': ['Random Forest', 'Bagging', 'AdaBoost', 'Voting'],
            'Accuracy': [rf_accuracy, bag_accuracy, ada_accuracy, voting_accuracy],
            'n_estimators': [n] * 4,
            'Pipelining': [pipelining] * 4
        })
        results = pd.concat([results, results_n])

    return results

# Run classifiers with and without pipelining
results_no_pipeline = run_classifiers(False)
results_with_pipeline = run_classifiers(True)

# Combine the results and store them in a CSV file
results = pd.concat([results_no_pipeline, results_with_pipeline])
results.to_csv('classifier_results.csv', index=False)

# Read the results from the CSV file into a DataFrame
results = pd.read_csv('classifier_results.csv')
```

```python
# Display the 5 overall best classifiers with the smallest n
top_5 = results.nlargest(5, ['Accuracy', 'n_estimators'])
top_5 = top_5.sort_values('n_estimators')
print("\nOverall 5 best classifiers:\n")
display(top_5[['Classifier', 'Accuracy', 'n_estimators', 'Pipelining']])
```
